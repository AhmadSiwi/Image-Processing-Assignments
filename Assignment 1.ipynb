```python
import cv2
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import skimage
from skimage import io
from skimage import filters
import numpy as np
```

```python
def Inkwell(image):
    i = image.copy()
    return cv2.cvtColor(i, cv2.COLOR_BGR2GRAY)

def Lily(image):
    i = image.copy()
    i[:,:,1] = np.maximum(127,i[:,:,1])
    return i

def Sutro(image):
    i = image.copy()
    j = i .copy()
    i[:,:,0] = np.minimum(j[:,:,0],j[:,:,1],j[:,:,2])
    i[:,:,1] = np.minimum(j[:,:,0],j[:,:,1],j[:,:,2])
    return i

def Gotham(image):
    i = image.copy()
    j = Inkwell(i)
    j = j-128
    j = np.multiply(j,(j<=127))
    j = j*2
    #To make dark pixels black and reserve the white ones as white
    return j

def Original(image):
    return image
#Original filter is mentioned in the photo in the slides so I can make it :P :"D

def insta_like(image, instafilter):
    if (instafilter == 'Lily'):
        image = Lily(image)
    if (instafilter == 'Inkwell'):
        image = Inkwell(image)
    if (instafilter == 'Original'):
        image = Original(image)
    if (instafilter == 'Sutro'):
        image = Sutro(image)
    if (instafilter == 'Gotham'):
        image = Gotham(image)
    return image
```

# 2.0 Input Image

```python
image = cv2.imread('index.jpeg', 1)
plt.axis("off")
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

# 2.1 Inkwell Filter

```python
i = insta_like(image, 'Inkwell')
plt.axis("off")
plt.imshow(cv2.cvtColor(i, cv2.COLOR_GRAY2RGB))
plt.show()
```

```python
j = insta_like(image, 'Gotham')
plt.axis("off")
plt.imshow(cv2.cvtColor(j, cv2.COLOR_GRAY2RGB))
plt.show()
```
